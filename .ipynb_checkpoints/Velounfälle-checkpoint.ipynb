```python
def get_bike_accidents(row):
    geodata = str(row[1].left) + ',' + str(row[1].bottom) + ',' + str(row[1].right) + ',' + str(row[1].top)
    hexagon_request = "https://api3.geo.admin.ch/rest/services/api/MapServer/identify?geometryType=esriGeometryEnvelope&geometry=" + str(geodata) + "&imageDisplay=500,600,96&mapExtent=548945.5,147956,549402,148103.5&tolerance=1&layers=all:ch.astra.unfaelle-personenschaeden_fahrraeder"
    try:
        df = pd.read_json(hexagon_request, orient='records', convert_axes=True)
        
    except:
        pass
    
    return df
```

```python
df = pd.concat([get_bike_accidents(row) for row in basel_grid.iterrows()])
```

```python
df.to_csv("bikes_raw.csv", index="False")
```

```python
def parse_bikes(index, row):
    content = pd.DataFrame(columns=['feature_id'])
    #print(row[1].results['featureId'])
    content.loc[index, 'feature_id'] = row['results']['featureId']
    content.loc[index, 'longitude'] = row['results']['geometry']['points'][0][0]
    content.loc[index, 'latitude'] = row['results']['geometry']['points'][0][1]
    content.loc[index, 'year'] = row['results']['attributes']['accidentyear']
    content.loc[index, 'canton'] = row['results']['attributes']['canton']
    content.loc[index, 'accidentday'] = row['results']['attributes']['accidentday_de']
    content.loc[index, 'accidenttype'] = row['results']['attributes']['accidenttype_de']
    content.loc[index, 'severity'] = row['results']['attributes']['severitycategory_de']
    content.loc[index, 'accidenttypecode'] = row['results']['attributes']['accidenttypecode']
    content.loc[index, 'roadtype'] = row['results']['attributes']['roadtype_de']
    return content
```

```python
def check_severity (row):
   if row['severity'] == 'Unfall mit Leichtverletzten' :
      return '1'
   if row['severity'] == 'Unfall mit Schwerverletzten' :
      return '2'
   if row['severity'] == 'Unfall mit Getöteten' :
      return '2'
```

```python
def check_weekday (row):
   if row['day'] == 'Samstag ' :
      return 'Wochenende'
   if row['day'] == 'Sonntag ' :
      return 'Wochenende'
   else :
      return 'Wochentag'
```

```python
def check_time (row):
    if row['start-time'] in [' 0500', ' 0600', ' 0700', ' 0800', ' 0900', ' 1000'] :
      return 'morgens'
    if row['start-time'] in [' 1100', ' 1200', ' 1300', ' 1400', ' 1500'] :
      return 'mittags'
    if row['start-time'] in [' 1600', ' 1700', ' 1800', ' 1900', ' 2000'] :
      return 'abends'
    else :
      return 'nachts'
```

```python
df_bikes_clean_split_2['severity_count'] = df_bikes_clean_split_2.apply(check_severity, axis=1)
```
